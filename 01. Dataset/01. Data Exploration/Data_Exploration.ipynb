## Load Data

### Subtask:
Load the `amzscout.csv` file into a pandas DataFrame.


```python
import pandas as pd

df = pd.read_csv('/content/amzscout.csv')
df.head()
```

## Display Data Info and Header




```python
print(f"DataFrame shape (rows, columns): {df.shape}")
print(f"DataFrame columns: {df.columns.tolist()}")
```

```python
df.head()
```

## Count Line Items by Product Category



```python
category_counts = df['Category'].value_counts()
print(category_counts)
```

## Display an Image


```python
import base64
from IPython.display import Image, display

# Find the first non-null image string in the 'Thumbnail Image' column
first_image_data = df['Thumbnail Image'].dropna().iloc[0]

# Remove the 'data:image/jpeg;base64,' prefix
if first_image_data.startswith('data:image/jpeg;base64,'):
    base64_string = first_image_data.split('data:image/jpeg;base64,')[1]
else:
    base64_string = first_image_data

# Decode the base64 string
image_bytes = base64.b64decode(base64_string)

# Display the image
display(Image(data=image_bytes))
print("Displayed the first available image from the 'Thumbnail Image' column.")
```

## Filter and Select Data



```python
target_categories = ['Sports & Outdoors', 'Home & Kitchen', 'Electronics']
filtered_df = df[df['Category'].isin(target_categories)][['Thumbnail Image', 'Product Name', 'URL', 'Category']]
print(f"Shape of the filtered DataFrame: {filtered_df.shape}")
print("First 5 rows of the filtered DataFrame:")
filtered_df.head()
```

## Rename Category Column


```python
filtered_df = filtered_df.rename(columns={'Category': 'Product Category'})
print("Columns after renaming:")
print(filtered_df.columns.tolist())
filtered_df.head()
```

## Add Product Classification


```python
filtered_df['Product Classification'] = 'New'
print("Columns after adding 'Product Classification':")
print(filtered_df.columns.tolist())
print("First 5 rows with new 'Product Classification' column:")
filtered_df.head()
```

## Sample and Save Data

```python
sampled_df = filtered_df.sample(n=100, random_state=42)
sampled_df.to_csv('New_images.csv', index=False)

print("Randomly sampled 100 rows and saved to 'New_images.csv'.")
print("Shape of sampled DataFrame:", sampled_df.shape)
print("First 5 rows of 'New_images.csv':")
print(sampled_df.head())
```
